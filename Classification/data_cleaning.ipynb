```python
import pandas as pd
import re
```

```python
lyrics_df = pd.read_csv('finalfinal.csv')
```

```python
lyrics_df
```

```python
#track - album - track_id - lyrics
lyrics_df = lyrics_df[['track', 'album', 'track_id', 'lyrics']]
lyrics_df
```

```python
train_emotion_df = pd.read_csv('train_labeled_lyrics.csv')
eval_emotion_df = pd.read_csv('eval_labeled_lyrics.csv')
```

```python
train_emotion_df.head()
```

```python
eval_emotion_df.head()
```

```python
emotion_df = pd.concat([train_emotion_df, eval_emotion_df], axis = 0, ignore_index = True)
emotion_df
```

```python
emotion_df = emotion_df[['track_id', 'emotion', 'intensity']]
emotion_df
```

```python
lyrics_emotion_df = pd.merge(lyrics_df, emotion_df, how='inner', on='track_id')
```

```python
lyrics_emotion_df
```

```python
lyrics_emotion_df.to_csv('lyrics_emotion_df')
```
