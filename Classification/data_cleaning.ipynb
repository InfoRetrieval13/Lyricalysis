```python
import pandas as pd
import re
import ast
import collections
import json
```

```python
lyrics_df = pd.read_csv('finalfinalfinalfinalfinal.csv')
```

```python
lyrics_df
```

```python
#track - album - track_id - lyrics
lyrics_df = lyrics_df[['artist', 'genre', 'track', 'duration', 'release_date', 'explicit', 'album', 'track_id', 'lyrics']]
lyrics_df
```

```python
train_emotion_df = pd.read_csv('train_labeled_lyrics.csv')
eval_emotion_df = pd.read_csv('eval_labeled_lyrics.csv')
```

```python
train_emotion_df.head()
```

```python
eval_emotion_df.head()
```

```python
emotion_df = pd.concat([train_emotion_df, eval_emotion_df], axis = 0, ignore_index = True)
emotion_df
```

```python
emotion_df = emotion_df[['track_id', 'emotion', 'intensity']]
emotion_df
```

```python
lyrics_emotion_df = pd.merge(lyrics_df, emotion_df, how='inner', on='track_id')
```

```python
lyrics_emotion_df
```

```python
#lyrics_emotion_df['genre'] = lyrics_emotion_df['genre'].apply(lambda x: ast.literal_eval(x))
lyrics_emotion_df['genre'] = lyrics_emotion_df['genre'].apply(lambda x: ", ".join(ast.literal_eval(x)))
```

```python
def duration_to_sec(x):
    x = x.split(' ')
    sec = int(x[0]) * 60 + int(x[2])
    return sec

lyrics_emotion_df['duration'] = lyrics_emotion_df['duration'].apply(lambda s: duration_to_sec(s))
```

```python
lyrics_emotion_df
```

```python
for i in range(0, len(lyrics_emotion_df)):
    if len(lyrics_emotion_df['release_date'].iloc[i]) < 10:
        lyrics_emotion_df['release_date'].iloc[i] = lyrics_emotion_df['release_date'].iloc[i] + '-01-01'
```

```python
artist_count = lyrics_emotion_df['artist'].value_counts().to_dict()
```

```python
lyrics_emotion_df['genre'].value_counts()
```

```python
def flatten(xss):
    return [x for xs in xss for x in xs]

genre_count = dict(collections.Counter(flatten(list(lyrics_emotion_df['genre']))))
```

```python
emotion_count = lyrics_emotion_df['emotion'].value_counts().to_dict()
```

```python
lyrics_emotion_df.to_csv('cleaned_lyrics_emotion_df.csv')
```

```python
with open("artist_count.json", "w") as outfile: 
    json.dump(artist_count, outfile)
    
with open("genre_count.json", "w") as outfile: 
    json.dump(genre_count, outfile)

with open("emotion_count.json", "w") as outfile: 
    json.dump(emotion_count, outfile)
```
